```python
#21 as z-score
import numpy as np
data = [10, 20, 30, 40, 50]
mean = np.mean(data)
print("Mean:", mean)
std_dev = np.std(data, ddof=1)
print("Standard Deviation:", std_dev)
z_scores = [(value - mean) / std_dev for value in data]
print("Z-scores:", z_scores)
x = 40
z_score = (x - mean) / std_dev
print("Z-score of 40:", z_score)
```

```python
#statistical inference
import numpy as np
import math
population = np.array([10,20,30,40,50,60,70])
N = len(population)
print("population size (N):", N)
mu = np.mean(population)
print("population mean:", mu)
sigma2 = np.var(population, ddof=0)
print("population variance:", sigma2)
sigma = np.sqrt(sigma2)
print("population standard deviation:", sigma)
sample = np.array([20,30,40,50])
n = len(sample)
print("sample size (n):", n)
x_bar = np.mean(sample)
print("sample mean:", x_bar)
s2 = np.var(sample, ddof=1)
print("sample variance:", s2)
s = np.sqrt(s2)
print("sample standard deviation (s):", s)
SE = sigma / math.sqrt(n)
print("standard error:", SE)
z = (x_bar - mu) / SE
print("z-score:", z)
```
