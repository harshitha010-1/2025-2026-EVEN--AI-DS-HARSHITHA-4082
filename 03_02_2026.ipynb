```python
s = {1, 2, 3, 4}

# add elements
s.add(5)
s.update([6, 7])

# remove elements
s.remove(2)
s.discard(10)

# set operations
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)
print(a & b)
print(a - b)

# properties
print(len(s))
print(3 in s)
```

```python
d = {"name": "John", "age": 20}

# add / update
d["age"] = 21
d["city"] = "Delhi"

# access
print(d["name"])
print(d.get("marks", "Not Found"))

# remove
del d["age"]
d.pop("city")

# loop
for key, value in d.items():
    print(key, ":", value)

# dictionary methods
print(d.keys())
print(d.values())
print(d.items())
```
